```python
match = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IPL_prediction/IPL_Matches_2008_2022.csv")
delivery = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IPL_prediction/IPL_Ball_by_Ball_2008_2022.csv")
```

```python
delivery = delivery.merge(match[['ID','Team1','Team2']],left_on='ID',right_on='ID')  #By me
```

```python
total_score_df = delivery.groupby(['ID','innings']).sum()['total_run'].reset_index()
total_score_df
```

```python
total_score_df = total_score_df[total_score_df['innings']==1]
total_score_df
```

```python
match_df = match.merge(total_score_df[['ID','total_run']],left_on='ID',right_on='ID')
match_df
```

```python
match_df.columns
```

```python
match_df['Team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Gujarat Titans',
    'Lucknow Super Giants'
    ]
```

```python
match_df["Team1"] = match_df["Team1"].str.replace('Delhi Daredevils','Delhi Capitals')
match_df["Team2"] = match_df["Team2"].str.replace('Delhi Daredevils','Delhi Capitals')

match_df["Team1"] = match_df["Team1"].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df["Team2"] = match_df["Team2"].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['Team1'].isin(teams)]
match_df = match_df[match_df['Team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df = match_df[['ID','City','WinningTeam','total_run']]
match_df
```

```python
delivery_df = match_df.merge(delivery,on='ID')
delivery_df
```

```python
delivery_df = delivery_df[delivery_df['innings']==2]
delivery_df
```

```python
delivery_df.shape
```

```python
delivery_df["current_score"] = delivery_df.groupby('ID').cumsum()['total_run_y']
```

```python
delivery_df
```

```python
delivery_df["runs_left"] = delivery_df["total_run_x"] - delivery_df["current_score"]
```

```python
delivery_df["overs"] = delivery_df["overs"] + 1
```

```python
delivery_df
```

```python
delivery_df["balls_left"] = 126 - (delivery_df['overs']*6 + delivery_df['ballnumber'])
```

```python
delivery_df
```

```python
# The columns we required batting team, bowling team, city, run left, ball left,
# wicket left, total_run_x, crr, rrr, result
```

```python
delivery_df['player_out'] = delivery_df['player_out'].fillna('0')
```

```python
delivery_df['player_out'] = delivery_df['player_out'].apply(lambda x: x if x == '0' else '1')
```

```python
delivery_df['player_out'] = delivery_df['player_out'].astype('int')
```

```python
wickets = delivery_df.groupby('ID').cumsum()['player_out'].values
```

```python
delivery_df['wickets_left'] = 10 - wickets
```

```python
delivery_df
```

```python
delivery_df.columns
```

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
delivery_df
```

```python
def result(row):
  return 1 if row['WinningTeam']==row['BattingTeam'] else 0
```

```python
delivery_df["result"] = delivery_df.apply(result,axis=1)
```

```python
delivery_df.head()
```

```python
delivery_df['BowlingTeam'] = delivery_df.apply(lambda row: row['Team2'] if row['BattingTeam'] == row['Team1'] else row['Team1'], axis=1)
```

```python
delivery_df
```

```python
final_df = delivery_df[[ 'BattingTeam', 'BowlingTeam', 'City', 'runs_left', 'balls_left',
 'wickets_left', 'total_run_x', 'crr', 'rrr', 'result']]
```

```python
final_df.head()
```

```python
final_df = final_df.sample(final_df.shape[0])
final_df.head()
```

```python
final_df["City"].unique()
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.sample()
```

```python
final_df = final_df[final_df['balls_left']!= 0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
x_train.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','BowlingTeam', 'City'])
], remainder='passthrough')
```

```python
pipe = Pipeline(steps = [
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[1]
```

```python
import pickle
pickle.dump(pipe,open('IPL_Preditor.pkl','wb'))
```
